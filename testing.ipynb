```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
d_parser = lambda x : pd.datetime.strptime (str(x), '%Y-%m-%d')
data = pd.read_csv ('../data/withDate.csv', parse_dates=['release_date'], date_parser=d_parser)
```

```python
data.drop(data.loc[data.release_date.isna()].index, inplace=True)
```

```python
data.to_csv('../data/withDate.csv')
```
